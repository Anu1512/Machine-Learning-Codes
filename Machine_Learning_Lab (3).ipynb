```python
import random
def generate_continuous_random_numbers(num_samples, mean, std_dev):
  random_numbers = [random.gauss(mean, std_dev) for _ in range(num_samples)]
  return random_numbers
num_samples = 1000
mean = 0
std_dev = 1
random_numbers = generate_continuous_random_numbers(num_samples, mean, std_dev)
sample_mean = sum(random_numbers) / num_samples
sample_std_dev = (sum((x - sample_mean) ** 2 for x in random_numbers) / num_samples) ** 0.5
print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std_dev)
```

Experiment-6

Using the iris dataset with known class labels compare the labeling error of the K-means and EM algorithms. Measure the error by assigning a class label to each example. Assume that the number of clusters are known.
http://www.ars.usda.gov/Services/docs.htm?docid=8964

```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.metrics import accuracy_score

# Load Iris dataset
iris = datasets.load_iris()
X = iris.data
y_true = iris.target

# Number of clusters (assuming you know this)
n_clusters = 3

# K-means clustering
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans_labels = kmeans.fit_predict(X)

# EM (Expectation-Maximization) clustering (Gaussian Mixture Model)
em = GaussianMixture(n_components=n_clusters, random_state=42)
em_labels = em.fit_predict(X)

# Function to relabel cluster assignments to match true class labels
def relabel_clusters(cluster_labels, true_labels):
    mapping = {}
    for cluster in np.unique(cluster_labels):
        if np.sum(cluster_labels == cluster) > 0:  # Check if cluster has samples assigned to it
            most_common_label = np.argmax(np.bincount(true_labels[cluster_labels == cluster]))
            mapping[cluster] = most_common_label
    relabeled = np.array([mapping.get(cluster, -1) for cluster in cluster_labels])  # Assign -1 for clusters without samples
    return relabeled

# Relabel clusters for K-means and EM
kmeans_labels_relabel = relabel_clusters(kmeans_labels, y_true)
em_labels_relabel = relabel_clusters(em_labels, y_true)

# Calculate accuracy (labeling error)
kmeans_accuracy = accuracy_score(y_true, kmeans_labels_relabel)
em_accuracy = accuracy_score(y_true, em_labels_relabel)

# Print results
print(f"Labeling error for K-means: {1 - kmeans_accuracy:.4f}")
print(f"Labeling error for EM: {1 - em_accuracy:.4f}")
```

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
# Load the Iris dataset
iris = load_iris()
X, y_true = iris.data, iris.target
# Set the number of clusters (assuming you know the correct number of classes in the Iris dataset) n_clusters = len(np.unique(y_true))
# Apply K-means clustering
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
y_kmeans = kmeans.fit_predict(X)
# Apply EM (Gaussian Mixture Model) clustering
gmm = GaussianMixture(n_components=n_clusters, random_state=42)
y_gmm = gmm.fit_predict(X)
# Evaluate labeling error using accuracy_score
accuracy_kmeans = accuracy_score(y_true, y_kmeans)
accuracy_gmm = accuracy_score(y_true, y_gmm)
# Display the results
print(f"Labeling error for K-means: {1 - accuracy_kmeans:.4f}")
print(f"Labeling error for EM (Gaussian Mixture): {1 - accuracy_gmm:.4f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Step 1: Download and load the data
data_url = "/content/nndb_flat.csv"
nutritional_data = pd.read_csv(data_url)

# Step 2: Preprocess the data
# Drop unnecessary columns
columns_to_drop = ['ID', 'FoodGroup', 'ShortDescrip', 'Descrip', 'CommonName', 'MfgName', 'ScientificName']
nutritional_data = nutritional_data.drop(columns=columns_to_drop)

# Replace 'No Data' with NaN
nutritional_data = nutritional_data.replace('No Data', np.nan)

# Fill missing values with zeros
nutritional_data = nutritional_data.fillna(0)

# Separate the nutritional values from the rest of the data
nutrients = nutritional_data.iloc[:, 1:]

# Standardize the data
scaler = StandardScaler()
nutrients_scaled = scaler.fit_transform(nutrients)

# Step 3: Apply Principal Component Analysis (PCA)
pca = PCA(n_components=2)
principal_components = pca.fit_transform(nutrients_scaled)
print(principal_components)
# Step 4: Visualize the results using a biplot
def biplot(score, coeff, labels=None):
    plt.figure(figsize=(10, 8))
    plt.scatter(score[:,0], score[:,1], alpha=0.5)
    for i in range(coeff.shape[0]):
        plt.arrow(0, 0, coeff[i,0], coeff[i,1], color='r', alpha=0.5)
        if labels is None:
            plt.text(coeff[i,0]*1.15, coeff[i,1]*1.15, nutrients.columns[i], color='g', ha='center', va='center')
        else:
            plt.text(coeff[i,0]*1.15, coeff[i,1]*1.15, labels[i], color='g', ha='center', va='center')
    plt.grid()
    plt.xlabel("PC1")
    plt.ylabel("PC2")
    plt.title("Biplot")

biplot(principal_components[:,0:2], np.transpose(pca.components_[0:2, :]), labels=nutrients.columns)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from hmmlearn import hmm
```

```python
base_dir ="https://github.com/natsunoyuki/Data_Science/blob/master/gold/gold/gold_price_usd.csv?raw=True"
data = pd.read_csv(base_dir)
# Convert the datetime from str to datetime object.
data["datetime"] = pd.to_datetime(data["datetime"])
# Determine the daily change in gold price.
data["gold_price_change"] = data["gold_price_usd"].diff()
# Restrict the data to later than 2008 Jan 01.
data = data[data["datetime"] >= pd.to_datetime("2008-01-01")]
# Plot the daily gold prices as well as the daily change.
plt.figure(figsize = (15, 10))
plt.subplot(2,1,1)
plt.plot(data["datetime"], data["gold_price_usd"])
plt.xlabel("datetime")
plt.ylabel("gold price (usd)")
plt.grid(True)
plt.subplot(2,1,2)
plt.plot(data["datetime"], data["gold_price_change"])
plt.xlabel("datetime")
```

```python
# Use the daily change in gold price as the observed measurements X.
X = data[["gold_price_change"]].values
# Build the HMM model and fit to the gold price change data.
model = hmm.GaussianHMM(n_components = 3, covariance_type = "diag",
n_iter = 50, random_state = 42)
model.fit(X)
# Predict the hidden states corresponding to observed X.
Z = model.predict(X)
states = pd.unique(Z)
```

Experiment-8

Implement the two classifiers logistic regression and relevance vector machine and compare their performance with accuracy, precision, and recall.
https://archive.ics.uci.edu/ml/datasets/human+activity+recognition+using+smartphones

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score
import os
import urllib.request
import zipfile

# Function to download and extract dataset
def download_dataset(url, save_path):
    if not os.path.exists(save_path):
        os.makedirs(save_path)
    zip_path = os.path.join(save_path, "UCI_HAR_Dataset.zip")
    urllib.request.urlretrieve(url, zip_path)
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(save_path)
    os.remove(zip_path)

# Download and extract the dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00240/UCI%20HAR%20Dataset.zip"
save_path = "/content/UCI HAR Dataset/UCI HAR Dataset"
download_dataset(url, save_path)

# Load the dataset
X_train = pd.read_csv("/content/UCI HAR Dataset/UCI HAR Dataset/train/X_train.txt", delim_whitespace=True, header=None)
y_train = pd.read_csv("/content/UCI HAR Dataset/UCI HAR Dataset/train/y_train.txt", header=None, squeeze=True)
X_test = pd.read_csv("/content/UCI HAR Dataset/UCI HAR Dataset/test/X_test.txt", delim_whitespace=True, header=None)
y_test = pd.read_csv("/content/UCI HAR Dataset/UCI HAR Dataset/test/y_test.txt", header=None, squeeze=True)

# Preprocess the data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Implement Logistic Regression classifier
logistic_model = LogisticRegression(max_iter=10000)
logistic_model.fit(X_train, y_train)
logistic_pred = logistic_model.predict(X_test)

# Implement RVM classifier using SGDClassifier with hinge loss
rvm_model = SGDClassifier(loss='hinge', alpha=1e-3, max_iter=1000, tol=1e-3, random_state=42)
rvm_model.fit(X_train, y_train)
rvm_pred = rvm_model.predict(X_test)

# Evaluate the classifiers
def evaluate_performance(y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, average='weighted', zero_division=0)
    recall = recall_score(y_true, y_pred, average='weighted', zero_division=0)
    return accuracy, precision, recall

logistic_accuracy, logistic_precision, logistic_recall = evaluate_performance(y_test, logistic_pred)
rvm_accuracy, rvm_precision, rvm_recall = evaluate_performance(y_test, rvm_pred)

# Print the results
print("Logistic Regression:")
print(f"Accuracy: {logistic_accuracy}")
print(f"Precision: {logistic_precision}")
print(f"Recall: {logistic_recall}")

print("\nRelevance Vector Machine (using SGDClassifier with hinge loss):")
print(f"Accuracy: {rvm_accuracy}")
print(f"Precision: {rvm_precision}")
print(f"Recall: {rvm_recall}")
```

Experiment-9

Implement a function that plots the average accuracy computed on a 5-fold cross validation vs different values of C for a linear SVM and different values of C and gamma on RBF SVM and report for each with average LOC curves.
https://archive.ics.uci.edu/ml/datasets/Breast+Cancer+Wisconsin+%28Diagnostic%
29

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score, StratifiedKFold, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import roc_curve, auc
import os
import urllib.request

# Function to load and preprocess the dataset
def load_data(url):
    # Download the dataset
    data = pd.read_csv(url, header=None)
    X = data.iloc[:, 2:].values
    y = data.iloc[:, 1].values
    y = np.where(y == 'M', 1, 0)  # Malignant: 1, Benign: 0

    # Preprocess the data
    scaler = StandardScaler()
    X = scaler.fit_transform(X)

    return X, y

# Function to plot average accuracy vs different values of C for linear SVM
def plot_linear_svm_accuracy(X, y):
    C_range = np.logspace(-2, 3, 10)
    accuracy_scores = []

    for C in C_range:
        clf = SVC(kernel='linear', C=C, probability=True)
        scores = cross_val_score(clf, X, y, cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42), scoring='accuracy')
        accuracy_scores.append(scores.mean())

    plt.figure()
    plt.plot(C_range, accuracy_scores, marker='o')
    plt.title('Average Accuracy vs C for Linear SVM')
    plt.xlabel('C')
    plt.ylabel('Average Accuracy')
    plt.xscale('log')
    plt.grid(True)
    plt.show()

# Function to plot average accuracy vs different values of C and gamma for RBF SVM
def plot_rbf_svm_accuracy(X, y):
    param_grid = {'C': np.logspace(-2, 3, 10),
                  'gamma': np.logspace(-3, 2, 10)}

    clf = SVC(kernel='rbf', probability=True)
    grid_search = GridSearchCV(clf, param_grid, cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42), scoring='accuracy', n_jobs=-1)
    grid_search.fit(X, y)

    # Extract average accuracy for each combination of C and gamma
    scores = grid_search.cv_results_['mean_test_score'].reshape(len(param_grid['C']), len(param_grid['gamma']))

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.imshow(scores, interpolation='nearest', cmap=plt.cm.hot)
    plt.title('Average Accuracy for RBF SVM')
    plt.xlabel('Gamma')
    plt.ylabel('C')
    plt.colorbar()
    plt.xticks(np.arange(len(param_grid['gamma'])), param_grid['gamma'], rotation=45)
    plt.yticks(np.arange(len(param_grid['C'])), param_grid['C'])
    plt.show()

    # Best parameters
    best_params = grid_search.best_params_
    print(f"Best parameters: {best_params}")

    # Plot average ROC curve for best parameters
    best_clf = SVC(kernel='rbf', C=best_params['C'], gamma=best_params['gamma'], probability=True)
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    tprs = []
    aucs = []
    mean_fpr = np.linspace(0, 1, 100)

    for train_index, test_index in skf.split(X, y):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        probas_ = best_clf.fit(X_train, y_train).predict_proba(X_test)
        fpr, tpr, thresholds = roc_curve(y_test, probas_[:, 1])
        tprs.append(np.interp(mean_fpr, fpr, tpr))
        tprs[-1][0] = 0.0
        roc_auc = auc(fpr, tpr)
        aucs.append(roc_auc)

    mean_tpr = np.mean(tprs, axis=0)
    mean_tpr[-1] = 1.0
    mean_auc = auc(mean_fpr, mean_tpr)

    plt.figure()
    plt.plot(mean_fpr, mean_tpr, color='b', label=f'Mean ROC (AUC = {mean_auc})')
    plt.plot([0, 1], [0, 1], linestyle='--', lw=2, color='r', label='Chance', alpha=.8)
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Average ROC Curve for RBF SVM')
    plt.legend(loc="lower right")
    plt.show()

# Main function
if __name__ == '__main__':
    url = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"
    X, y = load_data(url)

    plot_linear_svm_accuracy(X, y)
    plot_rbf_svm_accuracy(X, y)
```

Experiment-10

Implement the ensemble learning method using bagging method and voting classifier method for the breast cancer dataset in 8 and compute the average mean of accuracy for three classifiers.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import BaggingClassifier, VotingClassifier
from sklearn.metrics import accuracy_score

# Function to load and preprocess the dataset
def load_data(url):
    # Download the dataset
    data = pd.read_csv(url, header=None)
    X = data.iloc[:, 2:].values
    y = data.iloc[:, 1].values
    y = np.where(y == 'M', 1, 0)  # Malignant: 1, Benign: 0

    # Preprocess the data
    scaler = StandardScaler()
    X = scaler.fit_transform(X)

    return X, y

# Function to compute average mean accuracy using cross-validation
def compute_mean_accuracy(classifier, X, y):
    scores = cross_val_score(classifier, X, y, cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=42), scoring='accuracy')
    return np.mean(scores)

# Function to implement Bagging Classifier
def bagging_classifier(X, y):
    base_clf1 = DecisionTreeClassifier(random_state=42)
    base_clf2 = KNeighborsClassifier()
    base_clf3 = SVC(kernel='linear', probability=True)

    bagging_clf1 = BaggingClassifier(base_clf1, n_estimators=10, random_state=42)
    bagging_clf2 = BaggingClassifier(base_clf2, n_estimators=10, random_state=42)
    bagging_clf3 = BaggingClassifier(base_clf3, n_estimators=10, random_state=42)

    mean_accuracy_dt = compute_mean_accuracy(bagging_clf1, X, y)
    mean_accuracy_knn = compute_mean_accuracy(bagging_clf2, X, y)
    mean_accuracy_svc = compute_mean_accuracy(bagging_clf3, X, y)

    return mean_accuracy_dt, mean_accuracy_knn, mean_accuracy_svc

# Function to implement Voting Classifier
def voting_classifier(X, y):
    clf1 = DecisionTreeClassifier(random_state=42)
    clf2 = KNeighborsClassifier()
    clf3 = SVC(kernel='linear', probability=True)

    voting_clf = VotingClassifier(estimators=[('dt', clf1), ('knn', clf2), ('svc', clf3)], voting='hard')
    mean_accuracy_voting = compute_mean_accuracy(voting_clf, X, y)

    return mean_accuracy_voting

# Main function
if __name__ == '__main__':
    url = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"
    X, y = load_data(url)

    mean_accuracy_dt, mean_accuracy_knn, mean_accuracy_svc = bagging_classifier(X, y)
    mean_accuracy_voting = voting_classifier(X, y)

    print(f"Mean Accuracy (Bagging - Decision Tree): {mean_accuracy_dt}")
    print(f"Mean Accuracy (Bagging - KNN): {mean_accuracy_knn}")
    print(f"Mean Accuracy (Bagging - SVM): {mean_accuracy_svc}")
    print(f"Mean Accuracy (Voting Classifier): {mean_accuracy_voting}")
```
